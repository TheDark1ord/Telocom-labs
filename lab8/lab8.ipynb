## Упражнение 8.2
Для проверки преобразования Фурье гауссовой кривой напишем функцию, которая будет считать кривую нормального распределения и его БПФ

```python
def plot_gaussian(std):
    M = 32
    gaussian = scipy.signal.gaussian(M=M, std=std)
    gaussian /= sum(gaussian)

    plt.subplot(1, 2, 1)
    plt.plot(gaussian)
    decorate(xlabel='Time')

    fft_gaussian = np.fft.fft(gaussian)
    fft_rolled = np.roll(fft_gaussian, M//2)

    plt.subplot(1, 2, 2)
    plt.plot(np.abs(fft_rolled))
    decorate(xlabel='Frequency')
    plt.show()

plot_gaussian(2)
```

Попробуем различные значения std

```python
plot_gaussian(3)
```

```python
plot_gaussian(6)
```

Как видно из графиков, при увеличении std, ширина нормального распределения увеличивается, а ширина БПФ уменьшается
